```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the Excel file
df = pd.read_excel("main tasks with thier metrics.xlsx")

# Set seaborn style
sns.set(style="whitegrid")

# === Plot 1: Bar Chart - Number of Metrics per Task ===
plt.figure(figsize=(10, 6))
task_counts = df['Task'].value_counts()
sns.barplot(x=task_counts.values, y=task_counts.index, palette="Blues_d")
plt.title('Number of Metrics per Task')
plt.xlabel('Number of Metrics')
plt.ylabel('Task')
plt.tight_layout()
plt.show()

# === Plot 2: Pie Chart - Reference-based vs. Reference-free ===
ref_counts = df['Reference-based Free-based'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(ref_counts, labels=ref_counts.index, autopct='%1.1f%%', colors=["#66b3ff", "#ff9999"])
plt.title('Reference-based vs. Reference-free Metrics')
plt.tight_layout()
plt.show()

# === Plot 3: Stacked Bar Chart - Evaluation Type per Task ===
eval_type = df.groupby(['Task', 'Evaluation Type']).size().unstack(fill_value=0)
eval_type.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='Set2')
plt.title('Evaluation Type (Human vs. Automated) Across Tasks')
plt.ylabel('Number of Metrics')
plt.xlabel('Task')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# === Plot 4: Bar Chart - Distribution of Metrics by Evaluation Aspect ===
aspect_counts = df['Evaluation Aspect'].value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(x=aspect_counts.values, y=aspect_counts.index, palette="coolwarm")
plt.title('Distribution of Metrics by Evaluation Aspect')
plt.xlabel('Number of Metrics')
plt.ylabel('Evaluation Aspect')
plt.tight_layout()
plt.show()

# === Plot 5: Pie Chart - Implementation Complexity ===
complexity_counts = df['Implementation Complexity'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(complexity_counts, labels=complexity_counts.index, autopct='%1.1f%%', colors=sns.color_palette("pastel"))
plt.title('Metric Implementation Complexity')
plt.tight_layout()
plt.show()

# === Interesting Fact Highlight ===
# This is an annotation on the original chart. For now, just print the insight.
print(" Interesting Fact:\nDialogue Generation relies heavily on human-evaluated metrics, which stands out amid the field's push for automation.")
```

```python
# Load your Excel file
file_path = "main tasks with thier metrics1.xlsx"
df = pd.read_excel(file_path)

# Clean column names
df.columns = df.columns.str.strip().str.replace('\n', ' ').str.replace('-', '_')

df
```

```python
# Strip and normalize key fields
df['Task'] = df['Task'].str.strip()
df['Evaluation Type'] = df['Evaluation Type'].str.strip()
df['Evaluation Aspect'] = df['Evaluation Aspect'].str.strip()
df['Reference'] = df['Reference'].str.strip()

# -------------------------------
# 1. Count of Metrics per Task
plt.figure(figsize=(10, 6))
sns.countplot(data=df, y='Task', order=df['Task'].value_counts().index, palette='viridis')
plt.title('Number of Evaluation Metrics per NLP Task')
plt.xlabel('Count of Metrics')
plt.ylabel('NLP Task')
plt.tight_layout()
plt.savefig("metrics_per_task.png")
plt.show()

# -------------------------------
# 2. Distribution of Evaluation Types
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Evaluation Type', order=df['Evaluation Type'].value_counts().index, palette='Set2')
plt.title('Distribution of Evaluation Types')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("evaluation_type_distribution.png")
plt.show()

# -------------------------------
# 3. Reference-based vs. Reference-free
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Reference', palette='Set1')
plt.title('Reference-based vs. Reference-free Metrics')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("reference_based_distribution.png")
plt.show()

# -------------------------------
# 4. Metrics by Evaluation Aspect
plt.figure(figsize=(10, 6))
sns.countplot(data=df, y='Evaluation Aspect', order=df['Evaluation Aspect'].value_counts().index, palette='coolwarm')
plt.title('Evaluation Metrics by Aspect')
plt.xlabel('Count')
plt.ylabel('Evaluation Aspect')
plt.tight_layout()
plt.savefig("evaluation_aspect_count.png")
plt.show()
```
